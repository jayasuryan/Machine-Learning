##1 - Linear Regression
The file ***ex1data1.txt*** contains X-Y data.  The idea is to find the linear regression solution using Gradient Descent and Newton's method and compare it to the Normal Solution.

### Normal Solution
The idea is to find $\theta$ such that the cost $J(\theta) = \sum_{i=0}^{m-1} (y_i - \theta^T \bf{x_i})^2$ is minimized.  This is the linear least squares solution.  Please note that $\bf{x_i}$ is a vector that's equal in length to $\theta$ and has one more dimension (the intercept) than the original dimensions in the X data (1 in this case).  Y data has one dimension here (not necessary in the general case).

$J(\theta)$ can be expressed as $$J(\theta) = (Y - X \theta)^T (Y - X \theta) $$ where $Y$ is a $m\times1$ column vector containing all the Y data and $X$ is a $m\times2$ matrix containing input data as columns with the first column being all 1's for the intercept.

Minimizing $J(\theta)$ equals setting 

\begin{align}
&\frac{\partial}{\partial \theta} J(\theta) & = & 0\\
\implies & X^T X \theta & = &X^T Y \\
\implies & \theta & = & (X^T X)^{-1} X^T Y
\end{align}

This is the ***Normal Solution*** for Linear Regression

### Gradient Descent Method
For ***Gradient Descent***, we use the derivative of $J(\theta)$ to drive us towards the solution. For learning rate $\alpha$, we get a recursive equation for $\theta$ as follows.

\begin{align}
&\theta_{n+1} & = & \theta_n - \alpha \frac{\partial}{\partial \theta} J(\theta)\\
\implies &\theta_{n+1} & = & \theta_n + \alpha X^T(Y-X\theta)\\
\end{align}

### Newton's Method
For ***Newton's Method***, we use the Newton's method to find the solution of $\frac{\partial}{\partial \theta} J(\theta) = 0 $.  This is given by 
\begin{align}
&\theta_{n+1} & = & \theta_n - \frac{J^{'}(\theta)}{J^{''}(\theta)}\\
\implies &\theta_{n+1} & = & \theta_n + H^{-1}X^T(Y-X\theta)\\
\end{align}
where $H$ is the Hessian matrix and is given by $\left[\frac{\partial^2}{\partial \theta_i \partial \theta_j} J(\theta)\right]$ which for linear regression evaluates to $X^TX$.  Please note that Newton's method converges to the final value in one step and equates the Normal Solution in this case.

### LMS Algorithm
Least Mean Squares (LMS) is an iterative algorithm that converges to the LS solution using the gradient, but instead of looking at all the samples and using the mean value of the gradient for it's descent, it runs the algorithm on a per sample basis using $(y_i - \theta_i^T \bf{x_i}) \bf{x_i}$ as the gradient update to $\theta$ for the next sample.  Since, there's no averaging of the update going on here, the path that the algorithm takes is fairly noisy and might need multiple gear shifts of the update coefficient $\alpha$ to converge to the final solution.  How close you can get to the final solution depends on how noisy the LMS updates are.

### Where to use Linear Regression?
Linear regression is an optimal fitting solution if $y_n$ is a linear function of the observations $\bf{x_n}$, but when the observations are corrupted by Gaussian noise, i.e. $$y_n = \sum_{i=0}^{m-1} \alpha_i x_n(i) + n_n$$ where $n_n \sim \mathcal{N}(0, \sigma^2)$.

```python
def calculate_theta_update(Y, A, theta):
    E = Y - (A * (np.matrix(theta).transpose()))
    theta_adj = A.transpose() * E
    return theta_adj.getA1()
  
def calculate_cost(Y, A, theta):
    E = Y - (A * (np.matrix(theta).transpose()))
    return (E.transpose() * E).getA1()[0]

def calculate_theta_update_n(Y, A, theta):
    Hinv = np.linalg.inv(A.transpose() * A)
    E = Y - (A * (np.matrix(theta).transpose()))
    theta_adj = Hinv * A.transpose() * E
    return theta_adj.getA1()


close('all')
C = pandas.read_csv('ex1data1.txt', header=None, names='XY')
plt.figure(figsize=(10,8))
plt.grid(1)
plt.scatter(C['X'], C['Y'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('ex1data1.txt')

## Normal Solution
A = np.matrix([np.ones(len(C['X'])), C['X']])
A = A.transpose()
theta = np.linalg.pinv(A) * np.matrix([C['Y']]).transpose()
theta = theta.getA1()
print 'Normal Solution: y = {0}*x + {1}'.format(theta[1], theta[0])

# Test out
x = np.linspace(0, 25, 1000)
y = theta[0] + theta[1] * x
plt.plot(x, y, 'g', linewidth=2, label='Normal')

## Gradient Descent
alpha = 0.02 / len(C['X'])
num_iter = 1500
theta_list = [np.array([0, 0])]

A = np.matrix([np.ones(len(C['X'])), C['X']])
A = A.transpose()
Y = np.matrix([C['Y']]).transpose()

x = np.linspace(0, 25, 1000)
cost_list = []
for k in xrange(num_iter):
    theta = theta_list[-1]
    cost_list.append(calculate_cost(Y, A, theta))
    theta_adj = calculate_theta_update(Y, A, theta)
    theta_list.append(theta + alpha*theta_adj)
    
print 'Gradient Descent Result: y = {0}*x + {1}'.format(theta_list[-1][1], theta_list[-1][0])
plt.plot(x, theta_list[-1][0] + theta_list[-1][1] * x, 'r', linewidth=1.5, label='Grad Descent')  


##LMS 
alpha = 0.0005 #0.02 / len(C['X'])
theta_list_lms = [np.array([0, 0])]
Y = C['Y']
X = np.matrix([np.ones(len(C['X'])), C['X']]).T
num_iter = int(100000/len(C['Y']))

for i in xrange(num_iter):
    for k, y in enumerate(Y):
        theta = theta_list_lms[-1]
        theta_list_lms.append(theta + (alpha*(y - (X[k] * np.matrix(theta).T)) * X[k]).getA1())
    
print 'LMS Result: y = {0}*x + {1}'.format(theta_list_lms[-1][1], theta_list_lms[-1][0])
plt.plot(x, theta_list_lms[-1][0] + theta_list_lms[-1][1] * x, 'g', linewidth=1, label='LMS') 
plt.legend(loc=0)    


## Newton's method
alpha_n = 1.0
num_iter_n = 10
theta_list_n = [np.array([0, 0])]

A = np.matrix([np.ones(len(C['X'])), C['X']])
A = A.transpose()
Y = np.matrix([C['Y']]).transpose()

cost_list_n = []
for k in xrange(num_iter_n):
    theta = theta_list_n[-1]
    cost_list_n.append(calculate_cost(Y, A, theta))
    # Test out
    #plt.plot(x, theta[0] + theta[1] * x)  
    theta_adj = calculate_theta_update_n(Y, A, theta)
    theta_list_n.append(theta + alpha_n*theta_adj)
    
print 'Newton''s Method Result: y = {0}*x + {1}'.format(theta_list_n[-1][1], theta_list_n[-1][0])
plt.plot(x, theta_list_n[-1][0] + theta_list_n[-1][1] * x, 'k', linewidth=1, label='Newtons') 
plt.legend(loc=0)

## Plot Error surface
theta_0_list1 = np.linspace(-10, 10, 100)
theta_1_list1 = np.linspace(-1, 4, 100)

theta_0_list, theta_1_list = np.meshgrid(theta_0_list1, theta_1_list1)
cost = np.ndarray(shape=(len(theta_0_list[:,0]), len(theta_0_list[0,:])))
for k1, t0 in enumerate(theta_0_list1):
    for k2, t1 in enumerate(theta_1_list1):
        cost[k2, k1] = calculate_cost(Y, A, np.array([t0, t1]))
        
fig = plt.figure(figsize=(10,8))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(theta_0_list, theta_1_list, cost, rstride=1, cstride=1, cmap=cm.coolwarm,
        linewidth=0, antialiased=False)
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_xlabel(r'$\theta_0$')
ax.set_ylabel(r'$\theta_1$')
ax.set_zlabel(r'$J(\bf\theta)$')
plt.title('Error Surface')

plt.figure(figsize=(10,8))
levels = np.linspace(0, np.amax(cost)+1, 500)

norm = cm.colors.Normalize(vmax=np.amax(cost), vmin=0)
cmap = cm.gist_rainbow
plt.contour(theta_0_list, theta_1_list, cost, levels = levels)
plt.colorbar()
plt.plot([t[0] for t in theta_list], [t[1] for t in theta_list], 'k.', label='Grad Descent')
plt.plot([t[0] for t in theta_list_n], [t[1] for t in theta_list_n], '.', mfc='r', mec='r', label='Newton')
plt.plot([t[0] for t in theta_list_lms], [t[1] for t in theta_list_lms], '.', mfc='g', mec='g', label='LMS')
plt.xlabel(r'$\theta_0$')
plt.ylabel(r'$\theta_1$')
plt.legend(loc=0)
plt.title('Error Surface Contour')
```

## 2 - Logistic Regression
Logistic Regression model is used when observations $x_i, i = 0,\ldots,m-1$ are made on a Bernoulli Random variable $y \sim Ber(p)$ where $p$ is unknown and needs to be found. The logistic function (a.k.a the sigmoid function) is used to model the hypothesis function $h_{\theta}(x)$ that is used to find the ML estimate of $p$ from the observations $x_i$.  It is given by $$h_{\theta}(x) = \frac{1}{1 + e^{-\theta^Tx}}$$.  Note that $h_{\theta}(x)$ is bounded between 0 and 1, as $\theta^T x$ goes from $-\infty$ to $\infty$.

Now the likelihood can be calculated as function of $\theta$ as $$L(\theta) = \prod_{i=0}^{m-1} h_{\theta}(x_i)^{y_i}(1- h_{\theta}(x_i))^{(1-y_i)}$$.  In stead of maximizing the likelihood, we can maximize the log-likelihood $l(\theta)$ w.r.t $\theta$ and that should give us the same result.
\begin{align}
l(\theta) & = & \sum_{i=0}^{m-1} y_i \log{h_{\theta}(x_i)} + (1-y_i)\log{(1- h_{\theta}(x_i))} \\
\end{align}


###Newton's Method
To calculate $\arg\max_{\theta} l(\theta)$, we use the Newton's method to find the zero of its derivative w.r.t to $\theta$.

First, let's calculate $\frac{d h_{\theta}(x)}{d \theta}$.

\begin{align}
\frac{d h_{\theta}(x)}{d \theta} & = & \frac{e^{-\theta^T \bf{x}} } {\left( 1 + e^{-\theta^T \bf{x}} \right)^2} \bf{x}\\
 & = & (1 - h_{\theta}(x)) h_{\theta}(x) \bf{x} \\
\end{align}

\begin{align}
\frac{d}{d\theta} l(\theta) & = & \sum_{i=0}^{m-1} \left( \frac{y_i} {h_{\theta}(x_i)} - \frac{1-y_i} {1- h_{\theta}(x_i)}\right) \frac{d h_{\theta}(x_i)}{d \theta}\\
 & = & \sum_{i=0}^{m-1} \left( y_i - h_{\theta}(x_i) \right) \bf{x_i} \\
 & = & X^T (Y - H_{\theta}(X))
\end{align}

Where $Y$ is a $m\times1$ column vector containing the outputs and $H_{\theta}(X) = \left[\frac{1}{1+e^{\theta^T \bf{x_i}}}\right]$ is a $m\times1$ column vector of the hypothesis based on the observations. Please note that $\bf{x_i}$ and $\theta$ are $(n+1) \times 1$ column vectors where $n$ is the number of dimensions in the observations space (one additional dimension is for the intercept term). $X$ is $m\times(n+1)$ matrix $[\bf{x_i}^T]$.

The Hessian matrix $\bf{H} = [\frac{\partial^2 l(\theta)}{\partial \theta_j \partial \theta_k}]$ can be calculated as
\begin{align}
&\frac{\partial^2 l(\theta)}{\partial \theta_j \partial \theta_k} & = & - \sum_{i=0}^{m-1} h_{\theta}(x_i) (1-h_{\theta}(x_i)) x_i(j) x_i(k) \\
\implies & \bf{H} & = & -X^T \cdot diag\{H_{\theta}(X)\} \cdot (I - diag\{H_{\theta}(X)\}) \cdot X
\end{align}

Please note that $\bf{H}$ is negative definitive matrix, which points to the fact that the zero of the likelihood derivative is a global maximum. (Both the diagonal matrices have positive determinant because they contain logistic function outputs which all lie between 0 and 1.  Now $B = -X^T X$ is a negative definite matrix, since for every non-zero vector $a$, $a^T B a = (XA)^T XA = -\parallel XA \parallel^2 \lt 0$ 

The final Newton's method solution for $\theta^*$ is
$$\theta^* = \theta_0 - \bf{H}^{-1} X^T (Y - H_{\theta_0}(X)) $$

```python
# Logistic Regression
q1X = pandas.read_csv('q1x.dat', header=None, names='12', sep=' ', skipinitialspace=True)
q1Y = pandas.read_csv('q1y.dat', header=None, names='1', sep=' ', skipinitialspace=True)
theta = np.array([0, 0, 0])

Y = np.matrix(q1Y['1']).T
X = np.matrix([np.ones(len(q1X['1'])), q1X['1'], q1X['2']]).T

def calc_h_theta_x(X, theta):
    h_theta = []
    for k in xrange(len(X)):
        h_theta.append(1.0 / (1.0 + np.exp((X[k] * np.matrix(theta).T).getA1()[0])))

    h_theta = np.matrix(h_theta).T
    return h_theta
    
def calc_error_log(X, Y, theta):
    h_theta = calc_h_theta_x(X, theta)
    E = (Y - h_theta)
    return (E.T * E)

def calc_theta_adj_lr_gd(X, Y, theta):
    h_theta = calc_h_theta_x(X, theta)
    theta_adj = X.T * (Y - h_theta)
    return theta_adj.getA1()
    
def calc_theta_adj_lr_n(X, Y, theta):
    h_theta = calc_h_theta_x(X, theta)
    der_l_theta = X.T * (Y - h_theta)
    
    h1 = np.diagflat(h_theta)
    h2 = np.diagflat(1.0 - h_theta)
    der_der_l_theta = - (X.T * h1) * (h2 * X)
    H_inv = np.linalg.inv(der_der_l_theta)
    theta_adj = H_inv * der_l_theta
    return theta_adj.getA1()

# Gradient Descent - like
theta_list = [np.array([0, 0, 0])]
alpha = 0.2 / len(Y)
num_iter = 10
cost_list = []
for k in xrange(num_iter):
    theta = theta_list[-1]
    cost_list.append(calc_error_log(X, Y, theta).getA1()[0])
    theta_adj = calc_theta_adj_lr_gd(X, Y, theta)
    theta_list.append(theta - alpha_n*theta_adj)
    
# Newton's method    
num_iter_n = 20
theta_list_n = [np.array([0, 0, 0])]
alpha_n = 1.0
cost_list_n = []
for k in xrange(num_iter_n):
    theta = theta_list_n[-1]
    cost_list_n.append(calc_error_log(X, Y, theta).getA1()[0])
    theta_adj = calc_theta_adj_lr_n(X, Y, theta)
    theta_list_n.append(theta + alpha_n*theta_adj)

# Plots    
h_t_x_gd =  calc_h_theta_x(X, theta_list[-1])
h_t_x_n =  calc_h_theta_x(X, theta_list_n[-1])
plt.figure(figsize=(10,8))
plt.plot(Y, label='Raw Output')
plt.plot(h_t_x_gd, label='Log Reg Out - GD')
plt.plot(h_t_x_n, label='Log Reg Out - Newton')    
plt.grid(1)
plt.legend(loc=0)
plt.xlabel('samples')
plt.ylabel('Y')
plt.title('Logistic Regression')

# Plot 3-D scatter plot and the logistic regression boundary
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')
n = 100
mks = ['o', '*']
cls = ['r', 'b']
for k in xrange(len(q1Y['1'])):
    if q1Y['1'][k] == 0:
        ax.scatter(q1X['1'][k], q1X['2'][k], q1Y['1'][k], c=cls[0], marker=mks[0])
    else:
        ax.scatter(q1X['1'][k], q1X['2'][k], q1Y['1'][k], c=cls[1], marker=mks[1])

ax.set_xlabel(r'$X_1$')
ax.set_ylabel(r'$X_2$')
ax.set_zlabel(r'$Y$')

q1X_2_bndry = (- theta_list_n[-1][0] -  theta_list_n[-1][1]*q1X['1']) / theta_list[-1][2]

ax.plot(q1X['1'], q1X_2_bndry)
plt.figure(figsize=(10,8))
plt.plot(q1X['1'], q1X_2_bndry)
plt.grid(1)
plt.xlabel(r'$X_1$')
plt.ylabel(r'$X_2$')
plt.title(r'Input boundary corresponding to $h_{\theta}(x)$ = 0.5')
```

# 3. Weighted Least Squares Regression
Weighted linear squares try to minimize the cost $$J(\theta) = \sum_{i=0}^{m-1} w_i (y_i - \theta^T \bf{x_i})^2$$ where $m$ is the number of observations, $y_i$ are the target values and $\bf{x_i}$ are noisy observation vectors of dimension $n+1$, and $w_i$ are the weights assigned to each observation.  Please note that if all observations are uniformly weighted, then this reduces to the linear least squares problem.  

The cost function can be expressed in matrix form as
$$J(\theta) = (Y - X\theta)^T W (Y - X \theta)$$
where $Y$ is a $m\times1$ column vector of $y_i$, $X$ is a $m\times(n+1)$ matrix containing all $\bf{x_i}^T$ and $W$ is a $m\times m$ diagonal matrix containing $w_i$.

##Normal Solution
The normal solution for $\theta$ that minimizes $J(\theta)$ can be found by finding the solution to $\frac{dJ(\theta)}{d\theta} = 0 $.
\begin{align}
& \frac{dJ(\theta)}{d\theta} & = & 0 \\
=> & 2 X^T W X \theta & = & 2 X^T W Y\\
=> & \theta^* & = & (X^T W X)^{-1} X^T W Y
\end{align}

## How does WLS differ from LS?
As mentioned above, WLS weights each input differently.  Usually, the weights are selected proportional to the inverse of the variances of the observations.  What this means is that WLS is local linear regression fit close to the selected data point and the regression fit can vary from data point to datapoint, since the weights differ from data point to data point causing the value of $\theta^*$ to change across data points.

```python
# Weighted Linear Regression
q2Y = pandas.read_csv('q2y.dat', header=None, names='1', sep=' ', skipinitialspace=True)
q2X = pandas.read_csv('q2x.dat', header=None, names='1', sep=' ', skipinitialspace=True)
Y = np.matrix(q2Y['1']).T
X = np.matrix(q2X['1']).T
plt.figure(figsize=(10,8))
plt.scatter(X, Y)
plt.grid(1)
plt.xlabel('q2X')
plt.ylabel('q2Y')

X2 = np.matrix([np.ones(len(q2X['1'])), q2X['1']]).T
# Normal Solution
theta_ns = inv(X2.T * X2) * X2.T *Y
Y_ns = X2 * theta_ns
Y_ns = Y_ns.getA1()
plt.plot(X, Y_ns, 'g', label='Normal Solution')

#WLS regression
def run_WLS(X2, Y, T):
    Y_wls = []
    for k, x in enumerate(X2.T[1].getA1()):
        W = np.diagflat(np.exp(-(X2.T[1, k] - X2.T[1]).getA1()**2/(2*T*T) ) )
        theta = (inv(X2.T * W * X2) * X2.T * W * Y).getA1()
        Y_wls.append(theta[0] + theta[1]*x)
    return Y_wls

Y_wls_0p8 = run_WLS(X2, Y, 0.8)
Y_wls_0p1 = run_WLS(X2, Y, 0.1)
Y_wls_0p2 = run_WLS(X2, Y, 0.3)
Y_wls_2 = run_WLS(X2, Y, 2)
Y_wls_10 = run_WLS(X2, Y, 10)

plt.plot(X, Y_wls_0p8, 'r.', label=r"WLS Solution $\tau$=0.8")
plt.plot(X, Y_wls_0p1, 'g.', label=r"WLS Solution $\tau$=0.1")
plt.plot(X, Y_wls_0p2, 'k.', label=r"WLS Solution $\tau$=0.2")
plt.plot(X, Y_wls_2, 'c.', label=r"WLS Solution $\tau$=2")
plt.plot(X, Y_wls_10, 'y.', label=r"WLS Solution $\tau$=10")

plt.title('Weighted Least Squares')
plt.legend(loc=0)
```

As can be seen in the above plot, as the value of $\tau$ in the weights $w_i = e^{-(x-x_i)^2/2\tau^2}$ increases, the weights start approaching 1 for all $w_i$, which asymptotes to unweighted least squares solution.
